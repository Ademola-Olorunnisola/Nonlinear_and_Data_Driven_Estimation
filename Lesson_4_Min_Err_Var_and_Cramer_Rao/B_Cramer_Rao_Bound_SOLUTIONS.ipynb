# Generate sample time series data

Same as Linear Least Squares example, without the exponential.

```python
t = np.linspace(0, 3, 60)
x = np.array([[1, 1, 2]]).T
H = np.vstack([t, np.sin(t), np.cos(2*t)]).T
y = H@x
```

# Two sensors

```python
noise_mean = 0

noise_std_1 = 2
noise_std_2 = 0.5

v = []
v_std = []
noise_i = 1
n_chunks = 0
while len(v)<len(t):
    if noise_i == 1:
        noise_i = 2
        noise_std = noise_std_2
        chunk_size = 10
    else:
        noise_i = 1
        noise_std = noise_std_1
        chunk_size = 10
        
    v_std = np.hstack( (v_std, [noise_std]*chunk_size) ) 
    v = np.hstack( (v, np.random.normal(noise_mean, noise_std, chunk_size)) )
    n_chunks += 1

v_std = v_std[0:len(t)]
v = v[0:len(t)]

v_std = np.column_stack(v_std).T
v = np.column_stack(v).T

y_tilde = y+v
```

```python
n_chunks
```

# Plotting

```python
plot_tme(t, y, y_tilde)
```

# Candidate models

```python
model_1_H = np.vstack([t, np.sin(t), np.cos(2*t)]).T
```

# Minimum Variance Estimate: Gauss-Markov Theorem

$\mathbf{\hat{x}}=(H^{\top}R^{-1}H)^{-1}H^{\top}R^{-1}\mathbf{\tilde{y}}$

```python
R = np.diag( np.ravel(v_std)**2)
print(R)
```

### Model 1

```python
x_1_hat = inv(model_1_H.T@inv(R)@model_1_H)@model_1_H.T@inv(R)@y_tilde
```

```python
y_1_hat = model_1_H@x_1_hat
plot_tme(t, y, y_tilde, y_1_hat)
```

```python
x_error = x_1_hat - x
print(x_error)
```

# Covariance of x error

```python
P = x_error@x_error.T
print(P)
```

# Cramer Rao Bound

```python
CRB = inv(model_1_H.T@inv(R)@model_1_H)
print(CRB)
```

```python
plt.plot(np.diag(P), 'o', color='red')
plt.plot(np.diag(CRB), 'o', color='black')

plt.xlabel('parameter number')
plt.ylabel('error variance')
```

# Solution 3

```python
def do_mev_crb():
    t = np.linspace(0, 3, 60)
    x = np.array([[1, 1, 2]]).T
    H = np.vstack([t, np.sin(t), np.cos(2*t)]).T
    y = H@x
    
    noise_mean = 0
    
    noise_std_1 = 1
    noise_std_2 = 1
    
    v = []
    v_std = []
    noise_i = 1
    n_chunks = 0
    while len(v)<len(t):
        if noise_i == 1:
            noise_i = 2
            noise_std = noise_std_2
            chunk_size = 50
        else:
            noise_i = 1
            noise_std = noise_std_1
            chunk_size = 150
            
        v_std = np.hstack( (v_std, [noise_std]*chunk_size) ) 
        v = np.hstack( (v, np.random.normal(noise_mean, noise_std, chunk_size)) )
        n_chunks += 1
    
    v_std = v_std[0:len(t)]
    v = v[0:len(t)]
    
    v_std = np.column_stack(v_std).T
    v = np.column_stack(v).T
    
    y_tilde = y+v
    
    model_1_H = np.vstack([t, np.sin(t), np.cos(2*t)]).T
    
    R = np.diag( np.ravel(v_std)**2)
    
    x_1_hat = inv(model_1_H.T@inv(R)@model_1_H)@model_1_H.T@inv(R)@y_tilde
    
    x_error = x_1_hat - x
    
    P = x_error@x_error.T
    
    CRB = inv(model_1_H.T@inv(R)@model_1_H)
    
    return P, CRB
```

### Iterate 1000 times

```python
Ps, CRBs = [], []
for i in range(1000):
    P, CRB = do_mev_crb()
    Ps.append(P)
    CRBs.append(CRB)
```

### Plot the mean diagonals

```python
P_mean = np.mean(Ps, axis=0)
CRB_mean = np.mean(CRBs, axis=0)

plt.plot(np.diag(P_mean), 'o', color='red')
plt.plot(np.diag(CRB_mean), 'o', color='black')

plt.xlabel('parameter number')
plt.ylabel('error variance')
```
